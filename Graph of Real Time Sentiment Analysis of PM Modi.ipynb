```python
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation
%matplotlib notebook
```

```python
plt.style.use('fivethirtyeight')
```

```python
frame_len = 10000

fig = plt.figure(figsize=(20,15))

def animate(i):
    data = pd.read_csv('sentiment.csv')
    #y1 = data['Trump']
    #y2 = data['Obama']
    y1 = data['Modi']
    #y4 = data['Kamra']
    #y5 = data['Sanders']
    #y6=data['Biden']

    if len(y1)<=frame_len:
        plt.cla()
        #plt.plot(y1, label='Donald Trump')
        #plt.plot(y2, label='Obama')
        plt.plot(y1, label='Modi')
        #plt.plot(y4, label='Kamra')
        #plt.plot(y5, label='Sanders')
        #plt.plot(y6, label='Biden')
        
    else:
        plt.cla()
        #plt.plot(y1[-frame_len: ], label='Donald Trump')
        #plt.plot(y2[-frame_len: ], label='Obama')
        plt.plot(y1[-frame_len: ], label='Modi')
        #plt.plot(y4[-frame_len: ], label='Kamra')
        #plt.plot(y5[-frame_len: ], label='Sanders')
        #plt.plot(y6[-frame_len: ], label='Biden')
    
    plt.legend(loc='upper left')
    plt.tight_layout()


ani = FuncAnimation(plt.gcf(), animate, interval=1000)
```
