```python
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.animation import FuncAnimation
%matplotlib notebook
```

```python
plt.style.use('fivethirtyeight')
```

```python
frame_len = 10000

fig = plt.figure(figsize=(10,8))

def animate(i):
    data = pd.read_csv('sentiment.csv')
    y1 = data['Trump']
    y2 = data['Sanders']
    y3=data['Biden']

    if len(y1)<=frame_len:
        plt.cla()
        plt.plot(y1, label='Donald Trump')
        plt.plot(y2, label='Bernie Sanders')
        plt.plot(y3, label='Joe Biden')
        
    else:
        plt.cla()
        plt.plot(y1[-frame_len: ], label='Donald Trump')
        plt.plot(y2[-frame_len: ], label='Sanders')
        plt.plot(y3[-frame_len: ], label='Biden')
    
    plt.legend(loc='upper left')
    plt.tight_layout()


ani = FuncAnimation(plt.gcf(), animate, interval=1000)
```

### Here popularity of President Trump is very high comparitvely because this analysis was carried out at the time when President annouced stimulus package of 2 tr to fight Covid-19 Pandemic.
