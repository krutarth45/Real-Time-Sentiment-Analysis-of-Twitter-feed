```python
import tweepy
from tweepy import Stream
from tweepy import StreamListener
import json
from textblob import TextBlob
import re
import csv
import nltk
nltk.download('punkt')
```

```python
biden=0
sanders=0
trump=0


header_name = ['Biden','Sanders','Trump']
with open('sentiment.csv', 'w') as file:
    writer = csv.DictWriter(file, fieldnames=header_name)
    writer.writeheader()

class Listener(StreamListener):
    
    def on_data(self, data):
        raw_twitts = json.loads(data)
        try:
            tweets = raw_twitts['text']
            
            tweets = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweets).split())
            tweets = ' '.join(re.sub('RT',' ', tweets).split())
            
            blob = TextBlob(tweets.strip())
            
           
            global sanders
            global biden
            global trump
            
            
            bidenS=0
            sandersS=0
            trumpS=0
            
            for sent in blob.sentences:
                if "Trump" in sent and not "Sanders" in sent and not "Biden" in sent:
                    trumpS = trumpS + sent.sentiment.polarity
                elif "Biden" in sent and not "Sanders" in sent and not "Trump" in sent:
                    bidenS = bidenS + sent.sentiment.polarity
                elif "Sanders" in sent and not "Biden" in sent and not "Trump" in sent:
                    sandersS = sandersS + sent.sentiment.polarity
                
            
            
            biden = biden + bidenS
            sanders = sanders + sandersS
            trump = trump + trumpS
            
            with open('sentiment.csv', 'a') as file:
                writer = csv.DictWriter(file, fieldnames=header_name)
                info = {
                    'Biden': biden,
                    'Sanders': sanders,
                    'Trump' : trump
                }
                writer.writerow(info)
            
            print(tweets)
            print()
        except:
            print('Error got')
        
    def on_error(self, status):
        print(status)
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
```

```python
twitter_stream = Stream(auth, Listener())
twitter_stream.filter(track = ['Biden','Sanders','Trump'])
```
