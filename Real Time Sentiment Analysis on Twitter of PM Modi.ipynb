```python
import tweepy
from tweepy import Stream
from tweepy import StreamListener
import json
from textblob import TextBlob
import re
import csv
import nltk
nltk.download('punkt')
```

```python
modi=0
#kamra=0
#biden=0
#sanders=0
##trump=0
#obama=0

header_name = ['Modi']
with open('sentiment.csv', 'w') as file:
    writer = csv.DictWriter(file, fieldnames=header_name)
    writer.writeheader()

class Listener(StreamListener):
    
    def on_data(self, data):
        raw_twitts = json.loads(data)
        try:
            tweets = raw_twitts['text']
            
            tweets = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweets).split())
            tweets = ' '.join(re.sub('RT',' ', tweets).split())
            
            blob = TextBlob(tweets.strip())
            
           
            #global sanders
            global modi
           # global kamra
            #global biden
           # global obama
           # global trump
            
            
            bidenS=0
            sandersS=0
            modiS=0
            kamraS=0
            trumpS=0
            obamaS=0
            
            for sent in blob.sentences:
                if "Modi" in sent:
                    modiS = modiS + sent.sentiment.polarity
               # elif "Kamra" in sent:
                 #   kamraS = kamraS + sent.sentiment.polarity
               # elif "Trump" in sent and not "Sanders" in sent and not "Biden" in sent:
                #    trumpS = trumpS + sent.sentiment.polarity
               # elif "Biden" in sent and not "Sanders" in sent and not "Trump" in sent:
                #    bidenS = bidenS + sent.sentiment.polarity
                #elif "Sanders" in sent and not "Biden" in sent and not "Trump" in sent:
                 #   sandersS = sandersS + sent.sentiment.polarity
                #elif "Obama" in sent:
                 #   obamaS = obamaS + sent.sentiment.polarity
            
            
            modi = modi + modiS
           # kamra = kamra + kamraS
            #biden = biden + bidenS
            #sanders = sanders + sandersS
            #obama = obama + obamaS
            #trump = trump + trumpS
            
            with open('sentiment.csv', 'a') as file:
                writer = csv.DictWriter(file, fieldnames=header_name)
                info = {
                    'Modi' : modi,
                    #'Kamra' : kamra
                }
                writer.writerow(info)
            
            print(tweets)
            print()
        except:
            print('Error got')
        
    def on_error(self, status):
        print(status)
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
```

```python
twitter_stream = Stream(auth, Listener())
twitter_stream.filter(track = ['Modi'])
```
